## conver these text into vectore as we can convert each movie into  vector and we will findout the nearest movies nearest moves as similarities 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv  =CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
print(vectors.shape)
print(vectors[0])
```

```python
cv.get_feature_names_out()
```

## make similar words as similar  ex: loving ,love,loved =love

```python
import nltk
from nltk.stem.porter import  PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return  " ".join(y)
```

```python
data['tags'] = data['tags'].apply(stem)
```

```python
print(data['tags'])
```

```python
data.shape
```

## now we are find the distance of the 2 vecors in the term of cosine angle 


```python
from sklearn.metrics.pairwise import  cosine_similarity
similarity =cosine_similarity(vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
sorted(similarity[0],reverse=True)
```

```python
data[data["title"]=="Batman Begins"].index[0]
```

```python
def recommend(movie):
    movie_index = data[data["title"]==movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key= lambda x:x[1])[1:10]
    for i in movies_list:
        print(data.iloc[i[0]].title)
    
```

```python
recommend('Superman')
```
